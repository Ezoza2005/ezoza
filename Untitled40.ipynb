# **KNN-qo'shni algoritmi**

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("UAE_2010_20220307.csv")
df.sample(10)
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for column in df.columns:
    df[column] = le.fit_transform(df[column])
```

```python
df.sample(15)
```

```python
df.shape
```

```python
corr_matrix = df.corr().abs()
corr_matrix.style.background_gradient(cmap='viridis')
```

```python
X = df.drop('time', axis=1).values
y = df['time']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=12) # random_state bu fixserlash

# k-NN
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=15) # k-ni qiymati ya'ni qo'shni nuqtalar soni

knn.fit(X_train, y_train)
```

```python
y_pridect=knn.predict(X_test)
```

```python
y_test.to_numpy()
```

```python
y_pridect
```

```python
from sklearn.metrics import jaccard_score
jaccard_score(y_test, y_pridect, average='macro')
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(confusion_matrix(y_test, y_pridect), annot=True)
plt.xlabel("Bashorat qiymat")
plt.ylabel("Asl qiymatlar")
plt.show()
```

```python
confusion_matrix(y_test, y_pridect)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
precision = precision_score(y_test, y_pridect,average='macro')
recall = recall_score(y_test, y_pridect,average='macro')
f1 = f1_score(y_test, y_pridect,average='macro')
accuracy = accuracy_score(y_test, y_pridect)
print(f"{precision=}\n{recall=}\n{f1=}\n{accuracy=}")
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pridect))
```

```python
f1 = []
for k in range(1,30):
    knn = KNeighborsClassifier(n_neighbors=k) # k-ni qiymati
    knn.fit(X_train, y_train)
    y_predict = knn.predict(X_test)
    f1.append(f1_score(y_test, y_predict,average='macro'))

plt.figure(figsize=(10, 7))
plt.plot(range(1,30),f1)
plt.xticks(range(1,30))
plt.grid()
plt.show()
```
